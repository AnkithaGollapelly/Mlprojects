#### 4.4 Feature Wise Visualization
#### 4.4.1 GENDER COLUMN
- How is distribution of Gender ?
- Is gender has any impact on student's performance ?
#### UNIVARIATE ANALYSIS ( How is distribution of Gender ? )

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Ensure the gender counts are consistent and order is fixed
gender_counts = df['gender'].value_counts()
gender_order = gender_counts.index  # preserves order in value_counts()

# Create figure with 1 row, 2 columns
fig, ax = plt.subplots(1, 2, figsize=(20, 10))

# Bar plot with same order
sns.countplot(x='gender', data=df, palette='bright', ax=ax[0],
              saturation=0.95, order=gender_order)
for container in ax[0].containers:
    ax[0].bar_label(container, color='black', size=20)
ax[0].set_title('Gender Count', fontsize=18)

# Pie chart with same order
ax[1].pie(gender_counts[gender_order], labels=gender_order, 
          explode=[0, 0.1], autopct='%1.1f%%',
          shadow=True, colors=['#ff4d4d', '#ff8000'])
ax[1].set_title('Gender Distribution', fontsize=18)

plt.show()
```

#### BIVARIATE ANALYSIS ( Is gender has any impact on student's performance ? ) 

```python
gender_group = df.groupby('gender').mean(numeric_only=True)
```

```python
plt.figure(figsize=(10, 8))

X = ['Total Average','Math Average']


female_scores = [gender_group['average score'][0], gender_group['math score'][0]]
male_scores = [gender_group['average score'][1], gender_group['math score'][1]]

X_axis = np.arange(len(X))
  
plt.bar(X_axis - 0.2, male_scores, 0.4, label = 'Male')
plt.bar(X_axis + 0.2, female_scores, 0.4, label = 'Female')
  
plt.xticks(X_axis, X)
plt.ylabel("Marks")
plt.title("Total average v/s Math average marks of both the genders", fontweight='bold')
plt.legend()
plt.show()
```

#### Insights 
- On an average females have a better overall score than men.
- whereas males have scored higher in Maths.
